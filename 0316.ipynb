```python
import numpy as np
import matplotlib.pyplot as py
```

```python
x=0
y=0
position=[[],[]]
p=[]
for i in np.arange(0,1000):
    Xs=np.random.random()-0.5
    x=x+Xs
    Ys=np.random.random()-0.5
    y=y+Ys
    position[0].append(Xs)
    position[1].append(Ys)
    p.append([x,y])
py.subplot(2,1,1)
py.plot(p)
py.subplot(2,1,2)
py.hist(position[0])
```

```python
x=0
y=[]
for i in np.arange(0,100):
    x=x+np.random.random()-0.5
    y.append(x)
py.subplot(2,1,1)
py.hist(y,bins=50)
py.subplot(2,1,2)
py.plot(y)
```
